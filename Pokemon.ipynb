```python
# Import the necessary Python libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the given data in to the `source` variable.
source = pd.read_csv("Pokemon.csv")
```

```python
source.head(20)
```

```python
graph = source.query("Total > 600").sort_values("Total", ascending = False).head(5)
graph
```

```python
graph = graph.drop(["Total", "Number", "Generation"], axis = 1)
```

```python
graph.plot.bar(stacked = True)
```

```python
source.plot.hexbin(x = "Attack", y = "Defense", gridsize = 12)
```

```python
source.query("Generation == 1").head(50).drop(["Generation", "Number", "Sp_Atk", "Sp_Def", "Total"], axis = 1).plot.area(stacked = False)
```
