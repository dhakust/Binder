```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("input/data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Type1.value_counts()
```

```python
display(df.Gen.value_counts(normalize = True))
df.Gen.value_counts(normalize = True).plot.pie()
```

```python
df.Gen.value_counts().plot.bar()
```

```python
df.plot.scatter(x = "Attack", y = "Defense")
```

```python
df.plot.hexbin(x = "SAtk", y = "SDef", gridsize = 20)
```

```python
df.query("Gen == 1").drop(["Number", "Gen", "Total"], axis = 1).plot.area()
```

```python
sns.set(
    style   = "darkgrid",
    palette = "deep",
    font    = "Fira Code",
    context = "notebook")
```

```python
with sns.plotting_context("talk"):
    sns.relplot(
        x    = "Attack",
        y    = "Defense",
        col  = "Gen",
        hue  = "Type1",
        data = df.query("Gen == 1 or Gen == 6"))
```

```python
with sns.axes_style("whitegrid"):
    sns.jointplot(x = "SAtk", y = "SDef", data = df, kind = "hex", color = "seagreen")
```

```python
with sns.color_palette("colorblind"):
    sns.countplot(x = "Gen", data = df)
```

```python
sns.relplot(x = "Number", y = "Total", hue = "Type1", data = df.head(50), kind = "line")
```

```python
subset = (df.query("Legendary == True and Type1 == 'Psychic'").drop(["Number", "Total", "Gen"], axis = 1))
display(subset)
sns.relplot(
    x     = "Attack",
    y     = "Defense",
    size  = "SAtk",
    hue   = "SDef",
    data  = subset,
    sizes = (25, 200))
```
