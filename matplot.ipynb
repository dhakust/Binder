```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("input/data.csv")
```

```python
df.info()
```

```python
display(df.Gen.value_counts())
df.Gen.value_counts().plot.bar()
```

```python
display(df.Gen.value_counts(normalize = True))
df.Gen.value_counts(normalize = True).plot.pie()
```

```python
display(df.sort_values("Attack", ascending = False).head(5))
df.plot.scatter(x = "Attack", y = "Defense")
```

```python
display(df.sample(5, random_state = 5))
df.plot.hexbin(x = "SAtk", y = "SDef", gridsize = 15)
```

```python
sub = df.query("Gen == 1 and Legendary == False").drop(["Number", "Gen", "Total"], axis = 1)
display(sub.tail(5))
sub.plot.area()
```
